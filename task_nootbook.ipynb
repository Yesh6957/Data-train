```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid")
```

```python
# Load the coffee sales dataset
df = pd.read_csv("data/Flipkart_Mobiles.csv")

# Show first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'  # Supports more Unicode characters

# Bar Plot: Brand Distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Brand', order=df['Brand'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Mobile Brand Distribution")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("figures/barplot_brand.png")
plt.show()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'

# Box Plot: Selling Price by Brand
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='brand', y='selling_price')
plt.xticks(rotation=45)
plt.title("Selling Price Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Selling Price (INR)")
plt.tight_layout()
plt.savefig("figures/boxplot_price.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'

# KDE Plot: Rating Distribution
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='rating', fill=True, color='green')
plt.title("Rating Density Plot")
plt.xlabel("Rating")
plt.ylabel("Density")
plt.tight_layout()
plt.savefig("figures/kdeplot_rating.png")
plt.show()
```

```python
df['memory'] = pd.to_numeric(df['memory'], errors='coerce')
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='memory', order=sorted(df['memory'].dropna().unique()))
plt.title("Distribution of RAM Sizes")
plt.xlabel("RAM (GB)")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("figures/countplot_memory.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='selling_price', bins=30, kde=True, color='skyblue')
plt.title("Selling Price Distribution")
plt.xlabel("Price (INR)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("figures/histogram_price.png")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxenplot(data=df, x='brand', y='rating')
plt.xticks(rotation=45)
plt.title("Rating Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Rating")
plt.tight_layout()
plt.savefig("figures/boxenplot_rating_brand.png")
plt.show()
```

```python
# Remove non-numeric characters and convert to float
df['storage'] = df['storage'].str.extract(r'(\d+)').astype(float)
df['memory'] = df['memory'].astype(float)  # Assuming you've already cleaned this earlier
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'

# Select numeric columns for correlation
numeric_cols = ['selling_price', 'original_price', 'rating', 'memory', 'storage']
corr = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig("figures/heatmap_correlation.png")
plt.show()
```
